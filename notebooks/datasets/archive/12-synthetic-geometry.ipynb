```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Define the rectangle's dimensions
length = 24
width = 12

# Create a function to draw a 3D rectangle
def draw_3d_rectangle(vertices, ax):
    # Define the lines connecting the vertices to form the rectangle
    lines = [
        [vertices[0], vertices[1]],
        [vertices[1], vertices[2]],
        [vertices[2], vertices[3]],
        [vertices[3], vertices[0]],
    ]

    # Plot the lines
    for line in lines:
        ax.plot(*zip(*line), color='b')

# Create a function to rotate the 3D rectangle around the Y-axis
def rotate_3d_rectangle(vertices, angle_degrees):
    angle_radians = np.radians(angle_degrees)
    rotation_matrix = np.array([
        [np.cos(angle_radians), 0, np.sin(angle_radians)],
        [0, 1, 0],
        [-np.sin(angle_radians), 0, np.cos(angle_radians)]
    ])
    rotated_vertices = np.dot(vertices, rotation_matrix.T)
    return rotated_vertices

# Create a figure and 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Define the 3D vertices of the original rectangle
vertices = np.array([
    [-length / 2, -width / 2, 0],
    [length / 2, -width / 2, 0],
    [length / 2, width / 2, 0],
    [-length / 2, width / 2, 0],
])

# Draw the original 3D rectangle
draw_3d_rectangle(vertices, ax)

# Rotate the 3D rectangle by 30 degrees around the Y-axis
angle_degrees = 30
rotated_vertices = rotate_3d_rectangle(vertices, angle_degrees)

# Plot the rotated 3D rectangle
draw_3d_rectangle(rotated_vertices, ax)

# Set axis limits
ax.set_xlim(-length / 2 - 5, length / 2 + 5)
ax.set_ylim(-width / 2 - 5, width / 2 + 5)
ax.set_zlim(-width / 2 - 5, width / 2 + 5)

# Set aspect ratio to equal for proper visualization
ax.set_box_aspect([1, 1, 1])

# Enable interactive 3D plot
ax.view_init(elev=20, azim=30)

# Show the plot interactively
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.title('Interactive 3D Rotation of Rectangle')
plt.grid(True)
plt.show()
```

```python
import ipywidgets as widgets
from IPython.display import display
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Create a function to plot an interactive 3D plot
def plot_3d():
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    
    # Generate some example data
    x = np.linspace(-5, 5, 100)
    y = np.linspace(-5, 5, 100)
    X, Y = np.meshgrid(x, y)
    Z = np.sin(np.sqrt(X**2 + Y**2))
    
    # Plot the 3D surface
    ax.plot_surface(X, Y, Z, cmap='viridis')
    
    # Set labels and title
    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')
    ax.set_zlabel('Z-axis')
    plt.title('Interactive 3D Plot')

    plt.show()

# Create an interactive button
button = widgets.Button(description="Plot 3D")
button.on_click(lambda b: plot_3d())

# Display the button
display(button)
```

```python
import numpy as np

# Create a function to rotate the 3D rectangle around the Y-axis
def rotate_3d_rectangle(vertices, angle_degrees):
    angle_radians = np.radians(angle_degrees)
    rotation_matrix = np.array([
        [np.cos(angle_radians), 0, np.sin(angle_radians)],
        [0, 1, 0],
        [-np.sin(angle_radians), 0, np.cos(angle_radians)]
    ])
    rotated_vertices = np.dot(vertices, rotation_matrix.T)
    
    # Project the 3D vertices into the X-Y plane (2D)
    projected_vertices = rotated_vertices[:, :2]  # Discard the Z-coordinate
    
    return projected_vertices
```
